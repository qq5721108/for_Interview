# 一.配置爬虫生成类

```python
import requests
import json
import time
import pandas as pd

class TenAddress(object):
    def __init__(self, loc, ak):
        self.loc = loc
        self.ak = ak
        
    def url(self):
        url = 'https://apis.map.qq.com/ws/geocoder/v1/?location=' + self.loc + '&key=' + self.ak + '&get_poi=1'
            
        return url
    
    def get_data(self):
        try:
            url = self.url()
            print(url)
            js = requests.get(url).text
            data = json.loads(js)
            if data['status'] == 0:
                js = {}
                js['city'] = data['result']['address_component']['city']
                js['province'] = data['result']['address_component']['province']
                yield js
            else:
                raise RuntimeError('出现错误')
        except:
            print('出现错误！')
                        
```

```python
if __name__ == '__main__':
    
    gps_data = pd.read_excel(r'C:\Users\Trine\Desktop\my_data\epi_dig\epi_guangdong3.0_2.xlsx')
    ak = 'LVJBZ-G7Q3P-6O6DQ-LA64Q-S42BQ-ZCFCD'
    
    print('----------开始转换！----------')
    start_time = time.time()
    data = pd.DataFrame(columns=['city', 'province'])
    gps_data['city'] = 0
    gps_data['province'] = 0
    
    for index, item in gps_data.iterrows():
        gps = str(item['纬度']) + ',' + str(item['经度'])
        par = TenAddress(gps, ak)
        dt = par.get_data()
        df = pd.DataFrame(dt, columns=['city', 'province'])
        #data = data.append(df, ignore_index=True)
        if len(df) != 0:
            gps_data.loc[index, 'province'] = df.loc[0, 'province']
            gps_data.loc[index, 'city'] = df.loc[0, 'city']
            print(df)
        
    #gps_data['province'] = data['province']
    #gps_data['city'] = data['city']
    end_time = time.time()
    print('所有数据已经处理完毕，共耗时{:.2f}秒'.format(end_time - start_time))
```

# 二.根据city划分区域

```python
yuedong = ['潮州市','汕头市','汕尾市','揭阳市']
yuexi = ['湛江市','茂名市','阳江市']
yuebei = ['清远市','韶关市','河源市','梅州市','云浮市']
zhu_angle = ['广州市','深圳市','珠海市','佛山市','东莞市','中山市','江门市','惠州市', '肇庆市']
gps_data['area_now'] = '0'

for index, item in gps_data.iterrows():
    if item['province'] == '海南省':
        gps_data.loc[index, 'area_now'] = '海南'
        continue
    gps_data.loc[index, 'area_now'] = '粤西' if (item['city'] in yuexi) else '粤东' if (item['city'] in yuedong) else \
        '粤北' if (item['city'] in yuebei) else '珠三角' if (item['city'] in zhu_angle) else 0
    
```
