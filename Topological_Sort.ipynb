```python
from collections import deque

class DirectedGraphNode:
    def __init__(self, x):
        self.label = x;
        self.neighbors = []

class Solution:
    def topBFS(self, graph):
        """
        拓扑排序，广度优先
        
        Parms:
            graph, List<DirectedGraphNode> 图
        
        Returns:
            ans, list 拓扑排序后的list
        """
        
        #初始化记录度数组count
        count = {}
        for n in graph:
            count[n.label] = 0
        
        for n in graph:
            for j in n.neighbors:
                count[j.label] += 1
        
        ans = []
        deque = zeroDeque()
        
        for i in graph:
            if count[i.label] == 0:
                ans.append(i.label)
                zeroDeque.append(i)
        
        while zeroDeque:
            x = zeroDeque.leftpop()
            count[x.label] -= 1 ## -1表示已经遍历
            for i in x:
                count[i.label] -= 1
                if(i.label == 0):
                    zeroDeque.append(i)
                    ans.append(i)
        return ans
    
    def topDFS(self, graph):
        """
        拓扑排序 深度优先版

        Prams:
            graph, List<DirectedGraph> 图

        Returns;
            ans, list 拓扑排序后的list
        """
    
        #初始化记录dic
        count = {}
        for n in graph:
            count[n.label] = 0

        for node in graph:
            for i in node:
                count[node.label] += 1

        ans = []
        #深度优先遍历
        for node in graph:
            if(node.label == 0):
                dfs(node, count, ans)
        return ans
    
    def dfs(self, node, count, ans):
        """
        topDFS辅助函数

        Parms:
            node, DirectedGraph 图节点
            count, dic 节点度记录dic
            ans, list 需要返回的list
        """

        ans.append[node]
        count[node.label] -= 1

        for sub_node in node.neighbors:
            count[sub_node.label] -= 1 
            if(sub_node.label == 0):
                dfs(sub_node, count, ans)
```
